```python
import pandas as pd

# 读取你已经成功的那个文件
df = pd.read_csv('airlines.csv')

# 1. 查看每一列的数据类型和非空值数量
print("--- Data Info ---")
df.info()

# 2. 统计每一列分别有多少个空值
print("\n--- Missing Values ---")
print(df.isnull().sum())
```

```python
# 将 'Name' 替换为你第一步打印出来的实际名字（注意大小写）
# 比如改为 'name'
df_clean = df.dropna(subset=['airline_name']) 

# 统一国家名格式
# 同样，如果打印出来是 'country'，就把下面两行都改成小写
col_country = 'country' # 这里根据你的打印结果修改
df_clean[col_country] = df_clean[col_country].astype(str).str.strip()

print(f"Original rows: {len(df)}")
print(f"Cleaned rows: {len(df_clean)}")
```

```python
# 5. 统计每个国家的航司数量，并取前 10 名
# 请确认你的列名是 'Country' 还是 'country'
col_name = 'Country' if 'Country' in df_clean.columns else 'country'

top_10_countries = df_clean[col_name].value_counts().head(10)

print("Top 10 Countries with most Airlines:")
print(top_10_countries)
```

```python
import matplotlib.pyplot as plt

# 6. 绘制柱状图
plt.figure(figsize=(12, 6))
top_10_countries.plot(kind='bar', color='skyblue')

# 设置图表标题和标签（必须用英文）
plt.title('Top 10 Countries by Number of Airlines', fontsize=14)
plt.xlabel('Country Name', fontsize=12)
plt.ylabel('Count', fontsize=12)

# 自动调整布局防止文字重叠
plt.tight_layout()
plt.show()
```
